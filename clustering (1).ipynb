```python
features = ['Age', 'Income', 'Total_spending', 'NumWebPurchases', 'NumStorePurchases', 'NumWebVisitsMonth', 'Recency']
```

```python
x = df[features].copy()
```

```python
x
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
x_scaled = scaler.fit_transform(x)
```

```python
x_scaled
```

# Model Building 

```python
from sklearn.cluster import KMeans
```

```python
wcs = []
for i in range(2,10):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(x_scaled)
    wcs.append(kmeans.inertia_)
```

```python
wcs
```

```python
plt.plot(range(2,10),wcs,marker = 'o')
plt.title(" ELbow Methos For Optimal K ")
plt.xlabel('Number of Clusters')
plt.ylabel('wcs')
plt.show()
```

```python
kmeans= KMeans(n_clusters = 6)
df['Clusters']= kmeans.fit_predict(x_scaled)
```

```python
df.head()
```

```python
cluster_summary = df.groupby('Clusters')[features].mean()
```

```python
cluster_summary
```

```python
df['Clusters'].value_counts()
```

```python
from sklearn.decomposition import PCA 
pca = PCA(n_components=2)

pca_data = pca.fit_transform(x_scaled)
df['PCA1'], df['PCA2'] = pca_data[:,0],pca_data[:,1]
```

```python
pca_data
```

```python
sns.scatterplot(x = 'PCA1', y = 'PCA2', hue='Clusters', data = df , palette= 'Set1')
plt.title('Customer Segmentation (PCA)')
plt.show()
```
